### 1.2.7 Metric Evaluation

- MAE (Primary Metric): It is a good choice because its unit follows the original data unit. It is useful when the data contains outliers, as it is not affected by them. This will be the main focus for this Machine Learning model since the data contains outliers.
- MAPE (Secondary Metric): Useful for understanding the percentage of error.
- R-Square: Used to evaluate a model's performance. The closer it is to 1, the better the model fits the data. However, this does not necessarily mean the predictions will be accurate.
- RMSE: Places more weight on errors in outlier data, making it suitable for focusing on reducing prediction errors.

**Why MAE first?**

- In this case, we put MAE on the top priority because MAE measures the average absolute difference between predicted and actual prices. For example, if a model's MAE is SAR 1,500, it means that, on average, the predicted car prices have an error of about SAR 1,500 from the actual market prices. This metric is easier for users to understand compared to others like MSE or RMSE, which are based on squared differences.

- For Syarah.com, this means the model can remain accurate for the majority of cars without being affected by a few extreme-priced vehicles, resulting in more realistic predictions for premium customers.

The smaller the MAE and MAPE values, the more accurate the model will be in predicting car prices based on the limitations of the features used.

```python
def quantiles(dataframe = df ,columns="Price"):
    Q1 = dataframe[columns].quantile(0.25)
    Q2 = dataframe[columns].quantile(0.5)
    Q3 = dataframe[columns].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + IQR * 1.5 
    lower_bound = Q1 - IQR * 1.5
    return Q1,Q2,Q3,IQR,upper_bound,lower_bound
```

## 4.6 Data Correlation

We will see the correlation of data between one column and another, the data that can be correlated is integer data. Before we can make a correlation, we must check the distribution of data first.

**Method**

**DAgostino-Pearson Test (Recomended)**
1. Best for large datasets (>5000 rows).
2. Tests normality based on skewness and kurtosis.
3. More reliable than Shapiro-Wilk for large datasets.

```python
Ho = "We assume Data is Normally Distributed" # Ho indicates no difference, no effect, no relationship
Ha = "We assume Data is NOT Normally Distributed"

statistic,pvalues = normaltest(df['Price'])
print(f"Statistic: {statistic:.4f}, p-value: {pvalues:.4f}")

if pvalues >= 0.05:
    print(f"""P-Value = {pvalues} >= 0.05. Fail reject Ho
          {Ho}""")
else:
    print(f"""P-Value = {pvalues} < 0.05. Reject Ho
          {Ha}""")
```

Since the data is not normally distributed we will use correlation with the spearman method

```python
plt.figure(figsize=(10, 5))

list_column = [col for col in df.columns if df[col].dtype in ['float64', 'int64']]

corr = df[list_column].corr(method='spearman')

mask = np.triu(np.ones_like(corr, dtype=bool))

sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', vmin=-1, vmax=1, cmap='vlag')

plt.title("Spearman Correlation Heatmap")
plt.show()
```

```python
correlation = df['Price'].corr(df['Engine_Size'],method='spearman')

# conclusion
if correlation > 0:
    if 0 < correlation <= 0.2:
        print(f'''Correlation Score = {correlation}. very weak (doesnt have correlation) | Direction : Positive''')
    elif 0.2 < correlation <= 0.4:
        print(f'''Correlation Score  = {correlation}. weak | Direction : Positive''')
    elif 0.4 < correlation <= 0.6:
        print(f'''Correlation Score  = {correlation}. moderate | Direction : Positive''')
    elif 0.6 < correlation <= 0.9:
        print(f'''Correlation Score  = {correlation}. strong | Direction : Positive''')
    elif correlation <= 1 :
        print(f'''Correlation Score  = {correlation}. very strong | Direction : Positive''')
else:
    if -0.2 <= correlation < 0:
        print(f'''Correlation Score = {correlation}. very weak (doesnt have correlation) | Direction : Negative''')
    elif -0.4 <= correlation < -0.2:
        print(f'''Correlation Score  = {correlation}. weak | Direction : Negative''')
    elif -0.6 <= correlation <= -0.4:
        print(f'''Correlation Score  = {correlation}. moderate Direction : Negative''')
    elif -0.9 <= correlation <= 0.6:
        print(f'''Correlation Score  = {correlation}. strong | Direction : Negative''')
    elif correlation <= -1 :
        print(f'''Correlation Score  = {correlation}. very strong | Direction : Negative''')
```

```python
list_mean_rmse = []
list_std_rmse = []

list_mean_mae = []
list_std_mae = []

list_mean_mape = []
list_std_mape = []

list_mean_r2 = []
list_std_r2 = []

list_algo = [lr,ridge,bayesian_ridge,gaussian_process,tree,knn,rf,xgb,lgbm,cb,stacking,voting]
list_algo_name = ['LinearRegression','Ridge','BayesianRidge','GausianRidge','DecisionTreeRegressor','KNeighborsRegressor','RandomForest','XGBRegressor','LGBMRegressor','CatBooster','StackingRegressor','VotingRegressor']

scorer = ['neg_root_mean_squared_error',
          'neg_mean_absolute_percentage_error',
          'neg_mean_absolute_error',
          'r2']


for algo in list_algo:
    model = algo
    # pipeline
    pipe  = Pipeline([
        ('scaling', transformer),
        ('modeling', model)
    ])

    # cross validation
    cv_score = cross_validate(
        estimator=pipe,
        X = X_train,
        y = y_train,
        cv = 5,
        scoring = scorer
    )
    list_mean_rmse.append(-1*cv_score['test_neg_root_mean_squared_error'].mean())
    list_std_rmse.append(-1*cv_score['test_neg_root_mean_squared_error'].std())
    list_mean_mae.append(-1*cv_score['test_neg_mean_absolute_error'].mean())
    list_std_mae.append(-1*cv_score['test_neg_mean_absolute_error'].std())
    list_mean_mape.append(-1*cv_score['test_neg_mean_absolute_percentage_error'].mean())
    list_std_mape.append(-1*cv_score['test_neg_mean_absolute_percentage_error'].std())
    list_mean_r2.append(cv_score['test_r2'].mean())
    list_std_r2.append(cv_score['test_r2'].std())
```

```python
df_cv = pd.DataFrame()
df_cv['algo'] = list_algo_name

df_cv['mean_RMSE'] = list_mean_rmse
df_cv['std_RMSE'] = list_std_rmse

df_cv['mean_MAE'] = list_mean_mae
df_cv['std_MAE'] = list_std_mae

df_cv['mean_MAPE'] = list_mean_mape
df_cv['std_MAPE'] = list_std_mape

df_cv['mean_R2'] = list_mean_r2
df_cv['std_R2'] = list_std_r2


df_cv = df_cv.sort_values(by='mean_MAE')
df_cv
```
